```python
import os
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/autofl_base.csv")
print(df.head())
```

```python
target = "success"
features = [
    "total_messages",
    "function_calls",
    "valid_function_calls",
    "function_repetitions",
    "function_and_args_repetitions",
    "content_length",
]
```

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, test_size=0.2, random_state=42, stratify=df["success"])
print(len(df_train), len(df_test))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import roc_auc_score

X_train, y_train = df_train[features], df_train[target].astype(int)
X_test,  y_test  = df_test[features],  df_test[target].astype(int)

model = make_pipeline(StandardScaler(), LogisticRegression(max_iter=3000))
model.fit(X_train, y_train)

p_test = model.predict_proba(X_test)[:, 1]
auc_test = roc_auc_score(y_test, p_test)
print(f"Test AUC: {auc_test:.3f}")
```

```python
def assign_risk_zone(p):
    if p < 0.3:
        return "High Risk"
    elif p < 0.7:
        return "Uncertain"
    else:
        return "Safe"

df_test = df_test.copy()
df_test["success_prob"] = p_test
df_test["risk_zone"] = df_test["success_prob"].apply(assign_risk_zone)

print(df_test["risk_zone"].value_counts(normalize=True).round(3))
df_test[["bug_id","model","success_prob","risk_zone"]].head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
import xgboost as xgb


target = "success"
features = [
    "total_messages",
    "function_calls",
    "valid_function_calls",
    "function_repetitions",
    "function_and_args_repetitions",
    "content_length",
]

df_train, df_test = train_test_split(df, test_size=0.2, random_state=42, stratify=df[target])
X_train, y_train = df_train[features], df_train[target].astype(int)
X_test, y_test   = df_test[features],  df_test[target].astype(int)


model_xgb = xgb.XGBClassifier(
    eval_metric="logloss",
    n_estimators=400,
    learning_rate=0.05,
    max_depth=4,
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=42,
    use_label_encoder=False
)

model_xgb.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=False)


p_test = model_xgb.predict_proba(X_test)[:, 1]
auc_test = roc_auc_score(y_test, p_test)
print(f"Test AUC: {auc_test:.3f}")


def assign_risk_zone(p):
    if p < 0.3:
        return "High Risk"
    elif p < 0.7:
        return "Uncertain"
    else:
        return "Safe"

df_test = df_test.copy()
df_test["success_prob_xgb"] = p_test
df_test["risk_zone_xgb"] = df_test["success_prob_xgb"].apply(assign_risk_zone)

print(df_test["risk_zone_xgb"].value_counts(normalize=True).round(3))
df_test[["bug_id", "model", "success_prob_xgb", "risk_zone_xgb"]].head()
```

```python
from sklearn.calibration import CalibratedClassifierCV, calibration_curve
import matplotlib.pyplot as plt
import numpy as np

cal_xgb = CalibratedClassifierCV(model_xgb, method="isotonic", cv="prefit")
cal_xgb.fit(X_test, y_test)

p_test_cal = cal_xgb.predict_proba(X_test)[:, 1]
prob_true_raw, prob_pred_raw = calibration_curve(y_test, p_test, n_bins=10)
prob_true_cal, prob_pred_cal = calibration_curve(y_test, p_test_cal, n_bins=10)

plt.figure(figsize=(6, 5))
plt.plot(prob_pred_raw, prob_true_raw, "s-", label="Raw XGBoost")
plt.plot(prob_pred_cal, prob_true_cal, "o-", label="Calibrated (Isotonic)")
plt.plot([0, 1], [0, 1], "k--", label="Perfect calibration")
plt.xlabel("Predicted probability")
plt.ylabel("Observed success rate")
plt.title("Calibration Curve â€” XGBoost vs. Isotonic")
plt.legend()
plt.grid(True)
plt.show()

from sklearn.metrics import roc_auc_score

auc_test_raw = roc_auc_score(y_test, p_test)
auc_test_cal = roc_auc_score(y_test, p_test_cal)
print(f"AUC (raw): {auc_test_raw:.3f}")
print(f"AUC (calibrated): {auc_test_cal:.3f}")

def assign_risk_zone(p):
    if p < 0.3:
        return "High Risk"
    elif p < 0.7:
        return "Uncertain"
    else:
        return "Safe"

df_test["success_prob_cal"] = p_test_cal
df_test["risk_zone_cal"] = df_test["success_prob_cal"].apply(assign_risk_zone)

print(df_test["risk_zone_cal"].value_counts(normalize=True).round(3))
df_test[["bug_id", "model", "success_prob_cal", "risk_zone_cal"]].head()
```

```python
from sklearn.utils import resample

X_train, y_train = df_train[features], df_train[target].astype(int)
X_test,  y_test  = df_test[features],  df_test[target].astype(int)

n_bootstrap = 10
preds = []

for i in range(n_bootstrap):
    X_bs, y_bs = resample(X_train, y_train, random_state=42 + i)
    model_bs = xgb.XGBClassifier(
        eval_metric="logloss",
        n_estimators=300,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.9,
        colsample_bytree=0.9,
        random_state=42 + i,
        use_label_encoder=False
    )
    model_bs.fit(X_bs, y_bs)
    preds.append(model_bs.predict_proba(X_test)[:, 1])

preds = np.vstack(preds)

risk_mean = preds.mean(axis=0)
risk_std = preds.std(axis=0)

df_test = df_test.copy()
df_test["risk_mean"] = risk_mean
df_test["risk_std"] = risk_std

def assign_risk_zone_uncertain(mean, std):
    if mean < 0.3 and std < 0.1:
        return "High Risk (Certain)"
    elif mean < 0.3 and std >= 0.1:
        return "High Risk (Uncertain)"
    elif 0.3 <= mean < 0.7 and std >= 0.1:
        return "Uncertain"
    elif mean >= 0.7 and std < 0.1:
        return "Safe (Certain)"
    elif mean >= 0.7 and std >= 0.1:
        return "Safe (Uncertain)"
    else:
        return "Moderate"

df_test["risk_zone_uq"] = [
    assign_risk_zone_uncertain(m, s) for m, s in zip(df_test["risk_mean"], df_test["risk_std"])
]

zone_dist = df_test["risk_zone_uq"].value_counts(normalize=True).round(3)
print("\nRisk zone distribution (UQ):")
print(zone_dist)

ctab = pd.crosstab(df_test["risk_zone_uq"], df_test[target], normalize="index").rename(columns={0:"fail_rate",1:"success_rate"})
print("\nSuccess/fail rates within each UQ zone:")
print(ctab.round(3))

plt.figure(figsize=(6,4))
plt.scatter(df_test["risk_mean"], df_test["risk_std"], alpha=0.35, s=10)
plt.xlabel("Mean predicted success probability")
plt.ylabel("Prediction std (uncertainty)")
plt.title("Bootstrap Ensemble Uncertainty Map â€” autofl_base")
plt.grid(True)
plt.show()

zone_dist
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report


df["success_score"] = df["valid_function_calls"] / (df["function_calls"] + 1e-6)

df["success_zone_label"] = pd.qcut(df["success_score"], q=3, labels=["Low", "Moderate", "High"])

print(df["success_zone_label"].value_counts())

features = [
    "total_messages",
    "function_calls",
    "valid_function_calls",
    "function_repetitions",
    "function_and_args_repetitions",
    "content_length",
]

le = LabelEncoder()
y_encoded = le.fit_transform(df["success_zone_label"])

X_train, X_test, y_train, y_test = train_test_split(df[features], y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

model_multi = xgb.XGBClassifier(
    objective="multi:softprob",
    num_class=3,
    eval_metric="mlogloss",
    n_estimators=400,
    max_depth=4,
    learning_rate=0.05,
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=42,
    use_label_encoder=False
)
model_multi.fit(X_train, y_train)

p_test = model_multi.predict_proba(X_test)
y_pred = p_test.argmax(axis=1)
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

n_bootstrap = 10
preds = []

for i in range(n_bootstrap):
    X_bs, y_bs = resample(X_train, y_train, random_state=42 + i)
    model_bs = xgb.XGBClassifier(
        objective="multi:softprob",
        num_class=3,
        eval_metric="mlogloss",
        n_estimators=300,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.9,
        colsample_bytree=0.9,
        random_state=42 + i,
        use_label_encoder=False
    )
    model_bs.fit(X_bs, y_bs)
    preds.append(model_bs.predict_proba(X_test))

preds = np.stack(preds)
risk_mean = preds.mean(axis=0)
risk_std = preds.std(axis=0)

meta_features = pd.DataFrame({
    "mean_low": risk_mean[:, 0],
    "mean_mod": risk_mean[:, 1],
    "mean_high": risk_mean[:, 2],
    "std_low":  risk_std[:, 0],
    "std_mod":  risk_std[:, 1],
    "std_high": risk_std[:, 2],
})
y_meta = y_test

meta_model = LogisticRegression(multi_class="multinomial", max_iter=2000)
meta_model.fit(meta_features, y_meta)

meta_pred = meta_model.predict(meta_features)
acc = accuracy_score(y_meta, meta_pred)
print(f"Meta-model accuracy: {acc:.3f}")
print(pd.crosstab(le.inverse_transform(y_meta), le.inverse_transform(meta_pred), normalize="index"))
```
