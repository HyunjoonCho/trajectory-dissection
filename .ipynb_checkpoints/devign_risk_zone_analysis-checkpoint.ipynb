```python
import os
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/devign_base.csv")
print(df.head())
```

## General Analysis

```python
df_new = df.copy()
df_new["split_new"] = df_new["split"]

rng = np.random.RandomState(42)

for mdl, g in df_new.groupby("model"):
    val_idx = g[g["split"] == "valid"].index
    test_idx = g[g["split"] == "test"].index
    
    n_val_move = len(val_idx) // 2
    n_test_move = len(test_idx) // 2
    
    move_val = rng.choice(val_idx, size=n_val_move, replace=False)
    move_test = rng.choice(test_idx, size=n_test_move, replace=False)
    
    df_new.loc[move_val, "split"] = "train"
    df_new.loc[move_test, "split"] = "train"

df = df_new.copy()
```

```python
df_train = df_new[df_new['split']=='train'].copy()
df_val = df_new[df_new['split']=='valid'].copy()
df_test = df_new[df_new['split']=='test'].copy()

print(df_train.shape)
print(df_val.shape)
print(df_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import roc_auc_score

target = "success"
features = [
    "total_messages",
    "function_calls",
    "valid_function_calls",
    "function_repetitions",
    "function_and_args_repetitions"
]

X_train, y_train = df_train[features], df_train[target].astype(int)
X_val, y_val     = df_val[features], df_val[target].astype(int)
X_test, y_test   = df_test[features], df_test[target].astype(int)

model = make_pipeline(StandardScaler(), LogisticRegression(max_iter=3000))
model.fit(X_train, y_train)

p_val  = model.predict_proba(X_val)[:, 1]
p_test = model.predict_proba(X_test)[:, 1]

auc_val  = roc_auc_score(y_val, p_val)
auc_test = roc_auc_score(y_test, p_test)

auc_val, auc_test

### Results almost random
```

```python
from sklearn.calibration import calibration_curve
import matplotlib.pyplot as plt
import numpy as np

prob_true, prob_pred = calibration_curve(y_val, p_val, n_bins=10)

plt.figure(figsize=(6, 6))
plt.plot(prob_pred, prob_true, "s-", label="Logistic Regression")
plt.plot([0, 1], [0, 1], "k--", label="Perfect calibration")
plt.xlabel("Predicted success probability")
plt.ylabel("True success rate")
plt.title("Calibration Curve (Validation Set)")
plt.legend()
plt.grid(True)
plt.show()

def assign_risk_zone(p):
    if p < 0.3:
        return "High Risk"
    elif p < 0.7:
        return "Uncertain"
    else:
        return "Safe"

df_test["success_prob"] = p_test
df_test["risk_zone"] = df_test["success_prob"].apply(assign_risk_zone)

df_test["risk_zone"].value_counts(normalize=True).round(3)
```

```python
import xgboost as xgb
from sklearn.metrics import roc_auc_score

X_train, y_train = df_train[features], df_train[target].astype(int)
X_val, y_val     = df_val[features], df_val[target].astype(int)
X_test, y_test   = df_test[features], df_test[target].astype(int)

model_xgb = xgb.XGBClassifier(
    eval_metric="logloss",
    n_estimators=300,
    learning_rate=0.05,
    max_depth=4,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    use_label_encoder=False
)

model_xgb.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    verbose=False
)

p_val_xgb  = model_xgb.predict_proba(X_val)[:, 1]
p_test_xgb = model_xgb.predict_proba(X_test)[:, 1]

auc_val_xgb  = roc_auc_score(y_val, p_val_xgb)
auc_test_xgb = roc_auc_score(y_test, p_test_xgb)

print(f"Validation AUC: {auc_val_xgb:.3f}, Test AUC: {auc_test_xgb:.3f}")
```

```python
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import roc_auc_score

cal_model = CalibratedClassifierCV(base_estimator=model_xgb, method="isotonic", cv="prefit")
cal_model.fit(X_val, y_val)

p_test_cal = cal_model.predict_proba(X_test)[:, 1]

auc_test_cal = roc_auc_score(y_test, p_test_cal)
print(f"Calibrated Test AUC: {auc_test_cal:.3f}")

def assign_risk_zone(p):
    if p < 0.3:
        return "High Risk"
    elif p < 0.7:
        return "Uncertain"
    else:
        return "Safe"

df_test["success_prob_cal"] = p_test_cal
df_test["risk_zone_cal"] = df_test["success_prob_cal"].apply(assign_risk_zone)

print(df_test["risk_zone_cal"].value_counts(normalize=True).round(3))
```

```python
from sklearn.utils import resample
import numpy as np
import xgboost as xgb

n_bootstrap = 20
preds = []

for i in range(n_bootstrap):
    X_bs, y_bs = resample(X_train, y_train, random_state=42 + i)
    
    model_bs = xgb.XGBClassifier(
        eval_metric="logloss",
        n_estimators=300,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.9,
        colsample_bytree=0.9,
        random_state=42 + i,
        use_label_encoder=False
    )
    model_bs.fit(X_bs, y_bs)
    
    preds.append(model_bs.predict_proba(X_test)[:, 1])

preds = np.vstack(preds)

risk_mean = preds.mean(axis=0)
risk_std = preds.std(axis=0)

df_test["risk_mean"] = risk_mean
df_test["risk_std"] = risk_std
```

```python
def assign_risk_zone_uncertain(mean, std):
    if mean < 0.3 and std < 0.1:
        return "High Risk (Certain)"
    elif mean < 0.3 and std >= 0.1:
        return "High Risk (Uncertain)"
    elif 0.3 <= mean < 0.7 and std >= 0.1:
        return "Uncertain"
    elif mean >= 0.7 and std < 0.1:
        return "Safe (Certain)"
    elif mean >= 0.7 and std >= 0.1:
        return "Safe (Uncertain)"
    else:
        return "Moderate"

df_test["risk_zone_uq"] = [
    assign_risk_zone_uncertain(m, s) for m, s in zip(df_test["risk_mean"], df_test["risk_std"])
]

df_test["risk_zone_uq"].value_counts(normalize=True).round(3)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.scatter(df_test["risk_mean"], df_test["risk_std"], alpha=0.4, s=10)
plt.xlabel("Mean Predicted Success Probability")
plt.ylabel("Prediction Std (Uncertainty)")
plt.title("Bootstrap Ensemble Uncertainty Map")
plt.grid(True)
plt.show()
```

## Model 

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score
import xgboost as xgb
import matplotlib.pyplot as plt

df = pd.read_csv("data/devign_base.csv")

def safe_div(numer, denom):
    denom = denom.replace(0, np.nan)
    return (numer / denom).fillna(0.0)

df["valid_rate"] = safe_div(df["valid_function_calls"], df["function_calls"])
df["repetition_rate"] = safe_div(df["function_repetitions"], df["total_messages"])
df["function_ratio"] = safe_div(df["function_calls"], df["total_messages"])

df["valid_x_repetition"] = df["valid_rate"] * df["repetition_rate"]
df["valid_x_func_ratio"] = df["valid_rate"] * df["function_ratio"]
df["repetition_x_func_ratio"] = df["repetition_rate"] * df["function_ratio"]
df["log_total_messages"] = np.log1p(df["total_messages"])
df["log_function_calls"] = np.log1p(df["function_calls"])
df["log_content_length"] = np.log1p(df["content_length"].replace(-1, 0))

base_cols = [
    "total_messages","function_calls","valid_function_calls",
    "function_repetitions","function_and_args_repetitions","content_length"
]
ratio_cols = ["valid_rate","repetition_rate","function_ratio"]
inter_cols = ["valid_x_repetition","valid_x_func_ratio","repetition_x_func_ratio"]
log_cols = ["log_total_messages","log_function_calls","log_content_length"]

FEATURES = base_cols + ratio_cols + inter_cols + log_cols
TARGET = "success"

summary_rows = []

for mdl in sorted(df["model"].unique().tolist()):
    df_m = df[df["model"] == mdl].copy()
    df_train = df_m[df_m["split"] == "train"]
    df_val   = df_m[df_m["split"] == "valid"]
    df_test  = df_m[df_m["split"] == "test"]

    if len(df_train) == 0 or len(df_val) == 0 or len(df_test) == 0:
        continue

    X_train, y_train = df_train[FEATURES], df_train[TARGET].astype(int)
    X_val,   y_val   = df_val[FEATURES],   df_val[TARGET].astype(int)
    X_test,  y_test  = df_test[FEATURES],  df_test[TARGET].astype(int)

    lr = make_pipeline(StandardScaler(with_mean=False), LogisticRegression(max_iter=1000))
    lr.fit(X_train, y_train)
    p_val_lr  = lr.predict_proba(X_val)[:, 1]
    p_test_lr = lr.predict_proba(X_test)[:, 1]
    auc_val_lr  = roc_auc_score(y_val, p_val_lr)
    auc_test_lr = roc_auc_score(y_test, p_test_lr)

    summary_rows.append({
        "model_name": mdl,
        "algo": "LogisticRegression",
        "val_auc": auc_val_lr,
        "test_auc": auc_test_lr,
        "n_train": len(df_train),
        "n_val": len(df_val),
        "n_test": len(df_test),
    })

    xgb_clf = xgb.XGBClassifier(
        eval_metric="logloss",
        n_estimators=400,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.9,
        colsample_bytree=0.9,
        random_state=42,
        use_label_encoder=False
    )
    xgb_clf.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=False)
    p_val_xgb  = xgb_clf.predict_proba(X_val)[:, 1]
    p_test_xgb = xgb_clf.predict_proba(X_test)[:, 1]
    auc_val_xgb  = roc_auc_score(y_val, p_val_xgb)
    auc_test_xgb = roc_auc_score(y_test, p_test_xgb)

    summary_rows.append({
        "model_name": mdl,
        "algo": "XGBoost",
        "val_auc": auc_val_xgb,
        "test_auc": auc_test_xgb,
        "n_train": len(df_train),
        "n_val": len(df_val),
        "n_test": len(df_test),
    })

summary_df = pd.DataFrame(summary_rows).sort_values(["model_name","algo"])


xgb_only = summary_df[summary_df["algo"]=="XGBoost"].sort_values("test_auc", ascending=False)
plt.figure(figsize=(8,4))
plt.bar(xgb_only["model_name"], xgb_only["test_auc"])
plt.ylabel("Test AUC")
plt.title("XGBoost Test AUC by model")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

summary_df.head(10)
```
